# Books recommender system

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
books = pd.read_csv("books_data/books.csv", sep=";", on_bad_lines='skip', encoding='latin-1')
books.head()
```

```python
books.shape
```

```python
books.columns
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
```

```python
books.head()
```

```python
books.rename(columns={
    'Book-Title': 'title',
    'Book-Author': 'author',
    'Year-Of-Publication': 'year',
    'Publisher': 'publisher',
    'Image-URL-L': 'img_url'
}, inplace = True)
```

```python
books.head()
```

```python
books.info()
```

```python
users = pd.read_csv("books_data/users.csv", sep=";", on_bad_lines='skip', encoding='latin-1')
users.head()
```

```python
users.shape
```

```python
ratings = pd.read_csv("books_data/ratings.csv", sep=";", on_bad_lines='skip', encoding='latin-1')
ratings.head()
```

```python
ratings.shape
```

```python
users.rename(columns={
    'User-ID': 'user_id'
}, inplace = True)
```

```python
ratings.rename(columns={
    'User-ID': 'user_id',
    'Book-Rating': 'rating'
}, inplace = True)
```

```python
users.head()
```

```python
ratings.head()
```

```python
ratings['user_id'].unique().shape
```

```python
# Fusionner le jeu de donnees contenant les notes des livres et les livres
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
# Determiner le nombre de fois qu'un livre a ete note
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={
    'rating': 'num_of_rating',
}, inplace = True)
```

```python
final_rating = ratings_with_books.merge(num_rating, on='title')
```

```python
final_rating.head()
```

```python
# Filtrer pour ne garder que les livres qui ont plus de 50 notes
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_rating.shape
```

```python
# Create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
# Construire le model
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
import pickle
```

```python
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(books_name, open('artifacts/books_name.pkl', 'wb'))
pickle.dump(final_rating, open('artifacts/final_rating.pkl', 'wb'))
pickle.dump(book_pivot, open('artifacts/book_pivot.pkl', 'wb'))
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1), n_neighbors=6)
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for book in books:
            print(book)
```

```python
book_name = 'Harry Potter and the Chamber of Secrets (Book 2)'
recommend_book(book_name)
```
